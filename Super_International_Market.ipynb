# **What is the trend sales and provit over time?**

```python
segment_data.loc[:, 'Order Date'] = pd.to_datetime(segment_data['Order Date'])
segment_data.loc[:, 'Year'] = segment_data['Order Date'].dt.year
sales_profit_trend = segment_data.groupby('Year')[['Sales', 'Profit']].sum()
print(sales_profit_trend)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(sales_profit_trend['Sales'], label='Sales', color='#1f77b4', linewidth=2.5, marker='o')
plt.plot(sales_profit_trend['Profit'], label='Profit', color='#2ca02c', linewidth=2.5, marker='o')

plt.title('Sales and Profit Trend Over Time', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Amount($)', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--')
plt.legend(loc='upper left', fontsize=12)

plt.tight_layout()

plt.show()
```

**Sales and Profit Trend Report**

Grafik menunjukkan tren naik yang konsisten (garis biru menunjukkan penjualan dan garis hijau menunjukkan keuntungan) selama bertahun-tahun setiap 6 bulannya. Pertumbuhan penjualan menunjukkan peningkatan pendapatan, sementara pertumbuhan keutungan menunjukkan peningkatan profitabilitas. Fluktuasi dari tahun ke tahun menunjukkan dampak pasar dan operasional, tetapi keseluruhan menunjukkan kinerja bisnis yang kuat.

# Which regions and markets contribute the most to sales and profit?

```python
region_market_sales_profit = segment_data.groupby(['Region', 'Market'], observed=False)[['Sales', 'Profit']].sum()
top_region_sales= region_market_sales_profit.sort_values(by='Sales', ascending=False)
top_market_sales= region_market_sales_profit.sort_values(by='Profit', ascending=False)
```

```python
print('Top Regions by Sales')
top_region_sales[['Sales']].head()
```

```python
print("\nTop Markets by Profit:")
top_market_sales[['Profit']].head()
```

```python
top_regions_sales = region_market_sales_profit.sort_values(by='Sales', ascending=False).head()
top_markets_profit= region_market_sales_profit.sort_values(by='Profit', ascending=False).head()

fig, ax = plt.subplots(2, 1, figsize=(8, 8))

top_regions_sales[['Sales', 'Profit']].plot(kind='bar', stacked=True, ax=ax[0], color=['#1f77b4', '#2ca02c'])
ax[0].set_title('Top Regions by Sales and Profit', fontsize=14, color='darkblue')
ax[0].set_ylabel('Amount($)', fontsize=12)
ax[0].set_xlabel('Region', fontsize=12)

top_markets_profit[['Sales', 'Profit']].plot(kind='bar', stacked=True, ax=ax[1], color=['#1f77b4', '#2ca02c'])
ax[1].set_title('Top Markets by Sales and Profit', fontsize=14, color='darkblue')
ax[1].set_ylabel('Amount($)', fontsize=12)
ax[1].set_xlabel('Market', fontsize=12)

plt.tight_layout()

plt.show()
```

**Most Profitable Regions dan Markets**

Analisis ini menyoroti wilayah dan pasar yang memberikan kontribusi terbesar terhadap penjualan dan keuntungan. Diagram batang bertumpuk digunakan untuk membandingkan data penjualan dan keuntungan di berbagai wilayah serta pasar. Batang berwarna biru menunjukkan penjualan, sedangkan batang hijau menunjukkan keuntungan, sehingga memudahkan visualisasi wilayah dan pasar yang paling menguntungkan. Diagram ini dengan jelas mengidentifikasi wilayah dan pasar berkinerja tinggi, memberikan wawasan tentang faktor utama yang mendorong pertumbuhan penjualan dan keuntungan.

#Which cities and states are the most profitable?

```python
city_states_profit = segment_data.groupby(['City', 'State'])['Profit'].sum().reset_index()

top_city_states_profit = city_states_profit.sort_values(by='Profit', ascending=False).head()
top_city_states_profit
```

```python
top_city_states_profit = city_states_profit.sort_values(by='Profit', ascending=False).head()

fig, ax=plt.subplots(figsize=(10,6))

top_city_states_profit.plot(kind='bar', x='City', y='Profit', ax=ax, color='#2ca02c')

ax.set_title('Top Cities by Profit', fontsize=14, color='darkblue')
ax.set_ylabel('Profit ($)', fontsize=12)
ax.set_xlabel('City', fontsize=12)

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

**Most Profitable Cities and States**

Analisis ini mengidentifikasi kota dan negara bagian dengan kontribusi keuntungan terbesar. Diagram batang horizontal menampilkan pasangan kota dan negara bagian yang paling menguntungkan, dengan nilai keuntungan yang lebih tinggi terlihat dengan jelas. Kota-kota dengan keuntungan tertinggi menonjol, memungkinkan perbandingan profitabilitas secara cepat. Batang berwarna hijau menunjukkan total keuntungan, memberikan wawasan tentang area utama dalam kinerja bisnis.

#Which product are the most popular (highest quantity sold)?


```python
product_quantity = segment_data.groupby('Product Name')['Quantity'].sum().reset_index()

top_products = product_quantity.sort_values(by='Quantity', ascending=False).head()
top_products
```

```python
top_products = product_quantity.sort_values(by='Quantity', ascending=False).head()

fig, ax = plt. subplots(figsize=(10,6))

ax.pie(top_products['Quantity'], labels=top_products['Product Name'], autopct='%1.1f%%', colors=['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728', '#9467bd'])

ax.set_title('Top Products by Quantity Sold', fontsize=14, color='darkblue')

plt.tight_layout()
plt.show()
```

**Report: Most Popular Products by Quantity Sold**

Analisis ini menyoroti produk terlaris berdasarkan jumlah yang terjual. Diagram lingkaran digunakan untuk menampilkan proporsi penjualan setiap produk, di mana setiap produk direpresentasikan dengan warna yang berbeda. Segmen berwarna biru menunjukkan produk paling populer, sedangkan segmen berwarna hijau dan warna lainnya mewakili produk teratas berikutnya.

Diagram ini secara efektif menggambarkan pangsa setiap produk dalam total jumlah penjualan, memberikan gambaran yang jelas tentang produk paling populer dalam dataset. Representasi visual ini memungkinkan perbandingan popularitas produk berdasarkan volume penjualan dengan mudah.

# What is the sales and profit distribution acress different markets?

```python
market_sales_profit = segment_data.groupby('Market', observed=False)[['Sales', 'Profit']].sum().reset_index()
market_sales_profit
```

```python
fig, ax = plt.subplots(figsize=(10,6))

market_sales_profit.set_index('Market')[['Sales', 'Profit']].plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#2ca02c'])

ax.set_title('Sales and Profit Distribution Across Different Markets', fontsize=14, color='darkblue')
ax.set_ylabel('Amount ($)', fontsize=12)
ax.set_xlabel('Market', fontsize=12)

plt.tight_layout()
plt.show()
```

**Report: Sales and Profit Distribution Across Different Markets**

Analisis distribusi penjualan dan keuntungan di berbagai pasar divisualisasikan menggunakan diagram batang bertumpuk. Diagram ini menampilkan penjualan dalam warna biru dan keuntungan dalam warna hijau untuk setiap pasar. Dengan menumpuk nilai-nilai ini, perbandingan kontribusi total penjualan dan keuntungan di setiap pasar menjadi lebih mudah.

Diagram ini menyoroti pasar yang paling menguntungkan dan menghasilkan penjualan tertinggi. Pasar dengan batang yang lebih besar menunjukkan tingkat penjualan dan keuntungan yang lebih tinggi secara keseluruhan, sementara batang yang lebih kecil menunjukkan kontribusi yang lebih rendah. Visualisasi ini memberikan pemahaman yang jelas dan komparatif tentang kinerja pasar dalam hal penjualan dan profitabilitas.
